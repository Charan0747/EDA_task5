```python
#Exploratory Data Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing the Dataset
df = pd.read_csv(r"C:\Users\cherr\OneDrive\Documents\datasets\world_population.csv")
df
```

```python
#Setting the flaot values to 2 decimal points
pd.set_option('display.float_format',lambda x :'%.2f' % x)
```

```python
#Shows the Summary of the Dataset.
df.info()
```

```python
#Shows the Stastical Summary of the Dataset.
df.describe()
```

```python
#Calculation the Sum of the null values in each Column.
df.isnull().sum()
```

```python
#Getting the count of the uniques values.
df.nunique()
```

```python
#Sorting the Values in Descending order.
df.sort_values(by= "World Population Percentage",ascending = False).head(10)
```

```python
#Computing the Correlation between the Columns.
df.corr(numeric_only= True)
```

```python
#Showing the Correlation in the Heatmap.
plt.figure(figsize=(20,7))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
#Sorting the mean of the Population by Continent.
df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population",ascending=False)
```

```python
#Extracting the information of only Continent = Oceania.
df[df['Continent'].str.contains('Oceania')]
```

```python
#Getting mean of all the years and sorting them by 2022 Population and group up by Continent.
df2 = df.groupby('Continent')[['1970 Population',
        '1980 Population','1990 Population','2000 Population',
        '2010 Population','2015 Population','2020 Population',
        '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
#Extracting the column Names.
df.columns
```

```python
# Transposing the Coulmns 
df3 = df2.transpose()
df3
```

```python
#Plotting the df3.
df3.plot(figsize=(12,6))  # width=12 inches, height=6 inches
```

```python
#Plotting in boxplot to find the outliers and distribution of the data.
df.boxplot(figsize=(20,10))
```

```python
#To select the values of float.
df.select_dtypes(include='float')
```

```python
# Distribution of 2022 population
plt.figure(figsize=(10,5))
sns.histplot(df['2022 Population'], bins=30, kde=True)
plt.title("Distribution of 2022 Population")
plt.show()
```

```python
# Population by Continent (barplot for 2022)
plt.figure(figsize=(10,5))
sns.barplot(x='Continent', y='2022 Population', data=df, estimator=sum)
plt.title("Total Population by Continent (2022)")
plt.show()
```

```python
#Time series of top 5 populated countries
top5 = df.sort_values(by='2022 Population', ascending=False).head(5)
years = ['1970 Population','1980 Population','1990 Population','2000 Population',
         '2010 Population','2015 Population','2020 Population','2022 Population']
top5.set_index('Country')[years].T.plot(figsize=(12,6))
plt.title("Population Trends of Top 5 Countries")
plt.ylabel("Population")
plt.show()
```

```python
# Countries per continent (countplot)
plt.figure(figsize=(10,5))
sns.countplot(y='Continent', data=df, order=df['Continent'].value_counts().index)
plt.title("Number of Countries per Continent")
plt.show()
```
